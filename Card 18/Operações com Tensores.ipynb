```python
#squeeze, remove uma dimensão do tensor
a = torch.ones(2,1,10)
```

```python
#Tensor com 2 conjuntos, cada um com uma linha e 10 colunas
a
```

```python
print(f"shape antes do squeeze {a.shape}")
squeezed = a.squeeze(1)
print(f"shape depois do squeeze {squeezed.shape}")
```

```python
a = torch.ones(2,3,1)
print(f"este é o shape anetes do squeeze {a.shape}")
a
squeezed = a.squeeze(2)
print(f"essse é o shape depois do squeeze {squeezed.shape}")
```

```python
# unsqueeze, adiciona uma dimensão ao tensor
a = torch.ones(2,3)
print(f"shape antes do squeezing: {a.shape}")
squeezed = a.unsqueeze(2)
print(f"shape depois do squeeze {squeezed.shape}")
```

```python
#adicionando falsas dimensões ao tensor em índices específicos
print(a[None].shape)
print(a[:,None].shape)
print(a[:,:,None].shape)
a.shape
```

Operações com matrizes:

```python
x = torch.tensor([[1,2,3,4],[5,6,7,8]]) # sformato = 2x3
y = torch.tensor([[1,2,3],
                  [2,3,4],
                  [4,5,6],
                  [7,8,9]]) #formato = 4x3
```

```python
torch.matmul(x,y)
```

```python
# Concatenação
x = torch.randn(1,4,5)
z = torch.cat([x,x], axis = 1)
print('eixo concatenado 1:', x.shape, z.shape)
```

```python
#Permutação
x = torch.randn(3,20,10)
z = x.permute(2,0,1)
print('dimensões permutadas:',x.shape, z.shape)
```
