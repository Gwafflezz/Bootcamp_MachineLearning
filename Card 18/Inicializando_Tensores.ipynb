```python
import plotly.graph_objs as go
import plotly.offline as pyo

# criando um tensor tridimensional
a = torch.tensor([[[3,2],[3,4]],[[5,6],[7,8]]])

x_vals = []
y_vals = []
z_vals = []
for i in range(a.shape[0]):
  for j in range(a.shape[2]):
    for k in range(a.shape[1]):
      x_vals.append(i)
      y_vals.append(j)
      z_vals.append(k)

trace = go.Scatter3d(
    x=x_vals,
    y=y_vals,
    z=z_vals,
    mode='markers',
    marker=dict(
        size=5,
        color=a.flatten(),
        colorscale='Viridis',
        opacity=0.8
    )
)

layout = go.Layout(
    margin=dict(l=0,r=0,b=0,t=0),
     scene=dict(
        xaxis=dict(title='Dimensão 1'),
        yaxis=dict(title='Dimensão 2'),
        zaxis=dict(title='Dimensão 3')
    )
)

fig = go.Figure(data=[trace], layout=layout)
pyo.plot(fig, filename ='tensor.html')
```
