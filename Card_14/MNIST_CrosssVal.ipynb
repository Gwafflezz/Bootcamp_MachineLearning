```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from keras.utils import to_categorical
from keras.layers import BatchNormalization
import numpy as np
from sklearn.model_selection import StratifiedKFold
```

*Tratamento dos dados

```python

seed = 5
np.random.seed(seed)
#criação da variável
(X, y), (X_teste, y_teste) = mnist.load_data()
previsores = X.reshape(X.shape[0], 28, 28, 1) # transformando o formato dos dados
previsores = previsores.astype('float32') # define o tipo dos dados previsores
classe = to_categorical(y,10) # define as classes dos dados
```

```python
kfold = StratifiedKFold(n_splits = 5, shuffle = True, random_state = seed) # usa a função stratifiedkfold para criar os folds
resultados = []
```

```python
a = np.zeros(5)
b = np.zeros(shape = (classe.shape[0], 1))
```

```python
# definindo as porçoes de dados para treinamento e teste
for indice_treinamento, indice_teste in kfold.split(previsores,
                                                    np.zeros(shape = (classe.shape[0], 1))):
  print('Indices treinamento: ',indice_treinamento, 'Indice teste:', indice_teste)
```

**Criando a rede neural**


```python
classificador = Sequential()
classificador.add(Conv2D(filters=32,
                        kernel_size=(3,3),
                        input_shape=(28,28, 1),
                        activation='relu'))
classificador.add(MaxPooling2D(pool_size =(2,2)))
classificador.add(Flatten())
classificador.add(Dense(units = 128, activation ='relu'))
classificador.add(Dense(units = 10, activation = 'softmax'))
classificador.compile(loss = 'categorical_crossentropy', optimizer ='adam',
                      metrics = ['accuracy'])
classificador.fit(previsores[indice_treinamento],classe[indice_treinamento],
          batch_size = 128, epochs = 5)
precisao = classificador.evaluate(previsores[indice_teste], classe[indice_teste])


```

```python
resultados.append(precisao[1])
```

```python
media = resultados.mean()
media = sum(resultados/len(resultados))
```
