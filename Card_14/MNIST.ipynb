```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from keras.utils import to_categorical
from keras.layers import BatchNormalization


```

**Tratamento dos dados**

```python


(X_treinamento, y_treinamento), (X_teste, y_teste) = mnist.load_data()

plt.imshow(X_treinamento[0], cmap='gray')
plt.title('Classe: ' + str(y_treinamento[0]))

```

```python
previsores_treinamento = X_treinamento.reshape(X_treinamento.shape[0], 28, 28, 1)
previsores_teste = X_teste.reshape(X_teste.shape[0], 28, 28, 1)
```

```python
previsores_treinamento = previsores_treinamento.astype('float32')
previsores_teste = previsores_teste.astype('float32')

previsores_treinamento /= 255
previsores_teste /= 255

classe_treinamento = to_categorical(y_treinamento, 10)
classe_teste = to_categorical(y_teste, 10)
```

```python
classificador = Sequential()
classificador.add(Conv2D(filters=32,
                        kernel_size=(3,3),
                        input_shape=(28,28, 1),
                        activation='relu'))
classificador.add(BatchNormalization())  #melhoria1; Normaliza
classificador.add(MaxPooling2D(pool_size = (2,2)))
```

```python
classificador.add(Conv2D(filters=32,
                        kernel_size=(3,3),
                        activation='relu')) #melhoria 2; adição de uma segunda camada de convolução
classificador.add(BatchNormalization())
classificador.add(MaxPooling2D(pool_size = (2,2)))#melhoria 3
classificador.add(Flatten())
```

**Criação da rede neural densa:**

```python
classificador.add(Dense(units = 128, activation = 'relu'))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 128, activation = 'relu'))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 10, activation = 'softmax'))
```

```python
classificador.compile(loss='categorical_crossentropy',
                      optimizer = 'adam',
                      metrics = ['accuracy'])
```

```python
classificador.fit(previsores_treinamento,
                  classe_treinamento,
                  batch_size = 128, epochs = 5,
                  validation_data = (previsores_teste, classe_teste))
#Acho 5 um número muito baixo para treinar a rede, vamos ver o resultado
```

```python
resultado = classificador.evaluate(previsores_teste, classe_teste)
```
