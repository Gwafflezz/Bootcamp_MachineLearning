```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
outliers = np.random.normal(1000000, 100000, 500)
incomes = np.append(incomes, outliers)

import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show
```

```python
incomes.mean()
```

```python
def reject_outlier(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u+2 * s)]
    return filtered

filtered = reject_outlier(incomes)

plt.hist(filtered,90)
plt.show()

```

```python
np.mean(filtered)
```
