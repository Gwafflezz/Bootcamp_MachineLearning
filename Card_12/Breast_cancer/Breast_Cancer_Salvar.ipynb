```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
clasificador = Sequential()
clasificador.add(Dense(units = 8, activation= 'relu', kernel_initializer= 'normal', input_dim = 30))
clasificador.add(Dropout(0.2))
# mais camadas
clasificador.add(Dense(units = 8, activation= 'relu', kernel_initializer= 'normal',))
clasificador.add(Dropout(0.2))

clasificador.add(Dense(units= 1, activation = 'sigmoid'))
clasificador.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
clasificador.fit(previsores, classe, batch_size = 10, epochs = 100)
```

```python
classificador_json = clasificador.to_json()
with open('classificador_breast.json', 'w') as json_file:
    json_file.write(classificador_json)

clasificador.save_weights('classificador_breast.h5')
```
